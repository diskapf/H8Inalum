20 Game dengan Penjualan Terbaik

```python
df_game.sort_values(by='Global_Sales', ascending=True, inplace=True)
df_top20 = df_game['Global_Sales'].tail(20)
df_top20
```

```python
df_top20.plot(kind='barh', figsize=(12,12), color='steelblue')
plt.xlabel('Number of Copies Sold')
plt.title('Top 20 Best Selling Games in the World of All Time')
plt.show()
```

```python
df_game.info()
```

Publisher dengan Market Share Terbesar

```python
df_pub=df_game.groupby('Publisher', axis=0).sum()
print(type(df_game.groupby('Publisher', axis=0)))

df_pub.head()
```

```python
df_pub.sort_values(by='Global_Sales', ascending=True, inplace=True)
df_top_pub = df_pub['Global_Sales'].tail(10)
df_top_pub
```

```python
df_top_pub.plot(kin='pie',
               figsize=(5,6),
               autopct='%1.1f%%',
               startangle=90,
               shadow=True,
               )

plt.title('Top Games Publisher Market Share')
plt.axis('equal')

plt.show()
```

```python
df_year=df_game.groupby('Year', axis=0).sum()
print(type(df_game.groupby('Year', axis=0)))

df_year.head()
```

```python
import seaborn as sns
```

```python
sns.set_color_codes('pastel')
genre_feature = df_game['Genre'].unique()
sale = []

for genre in genre_feature:
    sale.append(df_game[df_game['Genre'] == genre]['Global_Sales'].sum())
    
    
fig,ax = plt.subplots(figsize=(8,6))
plt.title('Genre Sales',fontsize=15)
sns.barplot(x = genre_feature,y = sale , ax =ax,palette=sns.color_palette("PuBu", 10))
plt.ylabel('Total Sales',fontsize=15)
plt.xlabel('Genre Category',fontsize=15)
ticks = plt.setp(ax.get_xticklabels(),fontsize=12,rotation=90)
```
