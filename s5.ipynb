## Unexpected Missing Values

```python
df['OWN_OCCUPIED'] #mau mengconvert menjadi integer, kalau integer akan dijadikan missing values
```

```python
df['OWN_OCCUPIED'].isnull()
```

```python
cnt=0
for row in df['OWN_OCCUPIED'] :
    try:
        int(row)
        df.loc[cnt,'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
```
